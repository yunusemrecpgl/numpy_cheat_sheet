## Sifirdan array oluşturma

```python
np.zeros(10, dtype='int')
```

##### sifirdan olusan matris dondurur.

```python
np.ones((3,5), dtype='int')
```

##### birden olusan matris dondurur.

```python
np.full((3,5), 3)
```

##### ilk değer matris satir, ikinci değer matris sutun, ucuncu deger matris icindeki deger olarak matris dondurur.

```python
np.arange(0,31,3)
```

##### ilk değer baslangic, ikinci değer son deger, ucuncu artis miktari olan matris sonucu dondurur.

```python
np.linspace(0,1,10)
```

##### _linspace_ ilk değer baslangic, ikinci değer son deger, ucuncu deger sayısı olarak sonuc dondurur.

```python
np.random.normal(10, 4, (3,4))
```

##### _random.normal_ fonksiyonu ilk **değer ortalama**, ikinci değer **standart sapma**, ucuncu deger matris olarak rastgele sonuc dondurur.

```python
np.random.randint(0, 10, (3,3))
```

##### _random.randint_ fonksiyonu ilk değer baslangic, ikinci değer son deger, ucuncu deger matris olarak rastgele sonuc dondurur.
